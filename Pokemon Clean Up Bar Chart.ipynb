```python
import pandas as pd
```

```python
file = "300k.csv"
```

```python
Pokemon_DF = pd.read_csv(file)
Pokemon_DF.head()
```

```python
Pokemon_DF2 = Pokemon_DF.loc[:, ["pokemonId", "latitude", "longitude", "appearedLocalTime",
                                "appearedTimeOfDay", "appearedHour", "appearedMinute", "appearedDayOfWeek",
                                "appearedDay", "appearedMonth", "appearedYear","terrainType", "closeToWater",
                                "city","continent","weather","temperature", "windSpeed", "windBearing","pressure",
                                "weatherIcon","population_density", "urban", "suburban", "midurban", "rural", 
                                "gymDistanceKm", "class"]]
Pokemon_DF2
```

```python
columns = ["pokemonId", "latitude", "longitude", "appearedLocalTime",
                                "appearedTimeOfDay", "appearedHour", "appearedMinute", "appearedDayOfWeek",
                                "appearedDay", "appearedMonth", "appearedYear","terrainType", "closeToWater",
                                "city","continent","weather","temperature", "windSpeed", "windBearing","pressure",
                                "weatherIcon","population_density", "urban", "suburban", "midurban", "rural", 
                                "gymDistanceKm", "class"]

#America_Pokemon = Pokemon_DF2.loc[Pokemon_DF2["continent"] == "America", columns]
#Revist: Trying to grab three continents, possibly use Groupby 
America_Pokemon = Pokemon_DF2.loc[(Pokemon_DF2["continent"]== "America") | (Pokemon_DF2["continent"] == "America/Indiana"), columns]
America_Pokemon
```

```python
urban= (America_Pokemon["urban"]).value_counts().loc[True]
print(urban)
```

```python
suburban= (America_Pokemon["suburban"]).value_counts().loc[True]
print(suburban)
```

```python
midurban= (America_Pokemon["midurban"]).value_counts().loc[True]
print(midurban)
```

```python
rural= (America_Pokemon["rural"]).value_counts().loc[True]
print(rural)
```

```python
d = {'Demographic_count': [urban, suburban, midurban, rural], 'Demographic': ["Urban", "Suburban", "Midurban", "Rural"]}
df = pd.DataFrame(data=d)
df
```

```python
df.to_csv(r'urbsuburb.csv')
```
